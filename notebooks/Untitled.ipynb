## The 'hello world' of neural networks

How do x and y relate to each other in the following sequences ?

- X - -1, 0, 1, 2, 3, 4
- Y = -3, -1, 1, 3, 5, 7

The answer is `y = 2x - 1`

Let's write this using the tensorflow API <b>Keras</b>

A NN is basically a set of functions that can learn patterns !

```python
import keras
```

```python
# Simplest possible NN: one neuron
model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
# Dense: layer of connected neurons
# Sequential : successive layers
# nn might start with randomly guessing the x to y mapping
# how well it's doing is measured with the loss
# the loss is given to the optimizer, which then figures out the next guess
# logic : each guess should be better than the previous one
```

```python
# Let's represent the known data
import numpy as np
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

```python
# Training model, going through the training loop 500 times
model.fit(xs, ys, epochs=500)
```

```python
print(model.predict([10.0]))
```

#### Why are we not getting 19 ?

- training with only 6 points
- examples are linear, but no guarantee that for every x, y will be equal to 2x -1
- nn deal in probabilities
